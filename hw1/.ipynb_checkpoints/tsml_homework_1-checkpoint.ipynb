#### 1. Загрузка библиотек

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.tsa.api as smt

# Для проверки стационарности ряда будем использовать два метода (ADF и KPSS)
# https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html
# https://www.statsmodels.org/dev/examples/notebooks/generated/stationarity_detrending_adf_kpss.html
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss

import matplotlib.pyplot as plt
from scipy.stats import boxcox
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

#### 2. Загрузка датасетов

```python
sales_of_company_x = pd.read_csv("Series/monthly-sales-of-company-x-jan-6.csv")
robberies_in_boston = pd.read_csv("Series/monthly-boston-armed-robberies-j.csv")
airlines_passengers = pd.read_csv("Series/international-airline-passengers.csv")
mean_monthly_temp = pd.read_csv("Series/mean-monthly-air-temperature-deg.csv")
dowjones_closing = pd.read_csv("Series/weekly-closings-of-the-dowjones-.csv")
female_births = pd.read_csv("Series/daily-total-female-births-in-cal.csv")
```

#### 3. Визуальный анализ временных рядов

```python
all_series = {
    "Monthly sales of company X": sales_of_company_x["Count"],
    "Monthly Boston armed robberies": robberies_in_boston["Count"],
    "International airline passengers: monthly totals in thousands": airlines_passengers["Count"],
    "Mean monthly air temperature (Deg. F) Nottingham Castle": mean_monthly_temp["Deg"],
    "Weekly closings of the Dow-Jones industrial average": dowjones_closing["Close"],
    "Daily total female births in California": female_births["Count"]
}

with plt.style.context('bmh'):
    plt.figure(figsize=(16, 8))
    layout = (3, 2)
    for i, key in enumerate(all_series.keys()):
        x = i % 2
        y = int((i - x) / 2)
        
        ts_ax = plt.subplot2grid(layout, (y, x))
        all_series[key].plot(ax=ts_ax, color='blue')
        ts_ax.set_title(f'{i+1} {key}')
        
    plt.tight_layout()
```

#### 4. Определим базовые функции для отображения 

```python
# Определяем функцию, которая будет выводить результаты анализа 
# на стационарность / нестационарность временных рядов с использованием
# расширенного метода Дика-Фуллера ADF (Augmented Dickey-Fuller)
# и метода KPSS (Kwiatkowski-Phillips-Schmidt-Shin)
def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)

    print("\nResults of KPSS Test:")
    kpsstest = kpss(timeseries, regression="c", nlags="auto")
    kpss_output = pd.Series(
        kpsstest[0:3], index=["Test Statistic", "p-value", "Lags Used"]
    )
    for key, value in kpsstest[3].items():
        kpss_output["Critical Value (%s)" % key] = value
    print(kpss_output)
```

```python
# Эта функция для текущего временного ряда должна выводить статистику
# с помощью определенной выше функции, а также отображать графики 
# для ряда, его автокорреляции и кореллограмму
def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Временной ряд')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05, title='Автокорреляция')
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05, title='Коррелограмма')

        plt.tight_layout()
    return
```

#### 5. Анализ временных рядов и манипуляции с ними

Моментом достижения стационарности для каждого ряда будем считать уменьшение значения p-value ниже 0.05 для критерия ADF и, напротив, увеличение значения p-value выше 0.05 для критерия KPSS, поскольку нулевые гипотезы этих двух критериев являются противоположными. Кроме того, на каждом шаге будем принимать в расчет график изменения величин ряда и графики автокорреляции и коррелограмму.

##### 5.1. Ежемесячные продажи компании X (sales_of_company_x)

```python
# 5.1.1 - фиксируем данные
series_1 = sales_of_company_x["Count"]
```

```python
# 5.1.2 - предварительный анализ "сырых данных"
tsplot(series_1)
```

Мы видим, что у нас во-первых присутствует тренд, во-вторых явно увеличивается дисперсия. 
Также наблюдаются явные признаки сезонности.

```python
# 5.1.3 Вначале уберем влияние увеличивающейся дисперсии с помощью преобразования Бокса-Кокса
series_1_2 = boxcox(series_1, 0)
```

```python
tsplot(series_1_2)
```

Дисперсию мы победили, p-value слегка уменьшилось для ADF и пока не изменилось для KPSS. Пока ряд не стационарен.

```python
# 5.1.4 Уберем влияние тренда, используя дифференцирование
series_1_3 = series_1_2[1:] - series_1_2[:-1]
```

```python
tsplot(series_1_3)
```

Результат достигнут. Тест ADF дает основания отвегнуть гипотезу о нестационарности ряда (p-value 0.02 < 0.05), а тест KPSS перешел к состоянию, когда нельзя отвергнуть и гипотезу о стационарности (p-value 0.1 > 0.05).

```python
# 5.1.5 Тем не менее попробуем еще раз продифференцировать ряд:
series_1_4 = series_1_3[1:] - series_1_3[:-1]
```

```python
tsplot(series_1_4)
```

Мы видим, что статистические параметры стали еще лучше, также более компактным стал график автокорреляции. Коррелограмма сильно не изменилась.

#### Ежемесячная статистика по грабежам в Бостоне (robberies_in_boston)

```python
# 5.2.1 - фиксируем данные
series_2 = robberies_in_boston["Count"]
```

```python
# 5.2.2 - предварительный анализ "сырых данных"
tsplot(series_2)
```

В этом ряду явно прослеживается тренд. Не кажется логичным ожидать увидеть здесь какую-то сезонность. Также нет уверенности в изменении дисперсии, хотя амплитуда в начале ряда явно меньше амплитуды в конце.

```python
# 5.2.3 Начнем все-таки с преобразования Бокса-Кокса
series_2_2 = boxcox(series_2, 0)
tsplot(series_2_2)
```

Преобразование принесло свой эффект, амплитуда выровнялась. Дальше используем дифференцирование

```python
# 5.2.4 Попробуем дифференцировать данные
series_2_3 = series_2_2[1:] - series_2_2[:-1]
tsplot(series_2_3)
```

Результат достигнут. Оба теста + визуальный контроль однозначно подтверждают хороший уровень стационарности ряда.

#### Международные авиаперевозки пассажиров
Этот ряд был рассмотрен в качестве примера, сохраняю здесь код этого примера (добавив лишь использование критерия KPSS)

```python
series = pd.read_csv("Series/international-airline-passengers.csv")['Count']
```

```python
tsplot(series)
```

Как мы видим, и тест Дики-Фуллера и графики коррелограмм не отвергают гипотезу о нестационарности ряда. Для начала уберем изменение дисперсии при помощи преобразования Бокса-Кокса

```python
series = boxcox(series, 0)
```

```python
tsplot(series)
```

Нам удалось убрать размах дисперсии, но тест Дикки-Фуллера все еще не отвергает гипотезу о нестационарности ряда. По графику ряда видно наличие сильного тренда. Уберем его дифференцированием.

```python
series = series[1:] - series[:-1]
```

```python
tsplot(series)
```

Стало еще лучше, но по графику коррелограммы видно сильное влияние сезонности. Уберем ее

```python
series = series[12:] - series[:-12]
```

```python
tsplot(series)
```

Теперь тест Дики-Фуллера и графики коррелограмм отвергают гипотезу о нестационарности ряда!
